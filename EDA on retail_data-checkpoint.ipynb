```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
data = pd.read_csv("new_retail_data.csv")
df = pd.DataFrame(data)

print(df.head())
```

```python
df.shape
```

```python
df.info()
```

```python
# dropping unnecessary columns
df.drop(["Email","Phone","Address","Zipcode","Time"], axis=1, inplace=True)
```

```python
# Checking duplicate values in Transaction_ID column
duplicate_val = df["Transaction_ID"].duplicated().sum()
print(duplicate_val)

# Dropping duplicates on basis of Transaction_ID
df = df.drop_duplicates(subset=["Transaction_ID"])

print(df["Transaction_ID"].duplicated().sum())
```

```python
# Checking for null values

# Filling, replacing or dropping columns based on the null values
df = df.dropna(subset=["Transaction_ID"])
df = df.dropna(subset=["Customer_ID"])

# Non numerical Columns to fill null values
non_num_fill_nullVal = ["Name","City","State","Country","Gender","Income","Customer_Segment","Date","Year","Month","Product_Category","Product_Brand","Feedback","Shipping_Method","Payment_Method","Order_Status"]
df[non_num_fill_nullVal] = df[non_num_fill_nullVal].fillna(method="bfill")

# Numericall Columns to fill null values
num_fill_nullVal = ["Age", "Total_Purchases", "Amount", "Total_Amount", "Ratings"]
df[num_fill_nullVal] = df[num_fill_nullVal].fillna(np.floor(df[num_fill_nullVal].mean()))

print(df.isnull().sum())
```

```python
# Selecting specific columns and generating summary statistics
columns_to_stats = ["Age", "Total_Purchases", "Amount", "Total_Amount", "Ratings"]
summary_stats = df[columns_to_stats].describe()

print(summary_stats)
```

```python
# Transaction Patterns:
# What is the monthly and yearly trend in total transactions and revenue?

# Ensure 'Month' column is in the correct order
months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Grouping data on basis of months for each year 
months_2023 = df[df["Year"] == 2023].groupby("Month").agg({"Total_Amount":"sum"}).reindex(months_order, fill_value=0)
months_2024 = df[df["Year"] == 2024].groupby("Month").agg({"Total_Amount":"sum"}).reindex(months_order, fill_value=0)

# Grouping data for yearly sales
yearly_sales = df.groupby("Year")["Total_Amount"].sum()

fig, (fig1, fig2) = plt.subplots(1,2, figsize=(18,8))

#Comparing monthly revenue for each year
sns.lineplot(x= months_2023.index, y= months_2023["Total_Amount"], ls= "--", linewidth= 2, marker= "o", markersize= 8, label = "2023", ax= fig1)
sns.lineplot(x= months_2024.index, y= months_2024["Total_Amount"], ls= "--", linewidth= 2, marker= "o", markersize= 8, label = "2024", ax= fig1)
fig1.set_xlabel("Months", fontsize=14, fontweight= "bold", color= "black")
fig1.set_ylabel("Sales (in millions)", fontsize=14, fontweight= "bold", color= "black")
fig1.set_title("Total monthly sales ", fontsize=16, fontweight= "bold", color= "black")
fig1.grid(True, linestyle="--", alpha=0.7)

# Total Yearly sales
sns.barplot(x= yearly_sales.index, y= yearly_sales.values, ax= fig2, palette="viridis")
fig2.set_xlabel("Years", fontsize=14, fontweight= "bold", color= "black")
fig2.set_ylabel("Sales (in millions)", fontsize=14, fontweight= "bold", color= "black")
fig2.set_title("Total Yearly sales ", fontsize=16, fontweight= "bold", color= "black")
fig2.grid(axis='y', linestyle='-', alpha=0.5)

# Adding value labels on top of each bar
for p in fig2.patches:
    fig2.annotate(f"${p.get_height():,.0f}M", (p.get_x() + p.get_width() / 2., p.get_height()), 
                  ha='center', va='center', fontsize=10, color='black', xytext=(0, 10), 
                  textcoords='offset points')

plt.tight_layout()
plt.show()

# Conclusion
# The line plot says that there is a sudden rise in sales during the months of June, July and August in both year
```

```python
# What are the most and least popular products based on the total number of purchases?

# Filtering data
aggregating_products = df.groupby("Product_Type").agg({"Total_Purchases":"sum"}).sort_values(by="Total_Purchases", ascending=False)

# Most popular product based on the total number of purchases
most_pop = aggregating_products.head(1)
print("1) The most popular product based on total purchase is", most_pop.index[0], "with total number of purchase is", int(most_pop["Total_Purchases"].iloc[0]))

# Least popular product based on total number of purchases
least_pop = aggregating_products.tail(1)
print("2) The least popular product based on total purchase is", least_pop.index[0], "with total number of purchase is", int(least_pop["Total_Purchases"].iloc[0]))
```

```python
# # How do customer ratings and feedback scores correlate with the amount spent or the product categories purchased?

# Grouping or summarizing required data
requ_data = df[["Ratings", "Feedback", "Product_Category", "Total_Amount"]].copy()

# Converting  categorical or non-numerical data to numerical
requ_data["Feedback_Code"] = requ_data["Feedback"].astype("category").cat.codes
requ_data["Product_Category_Code"]= requ_data["Product_Category"].astype("category").cat.codes

# Updated columns to correlate
requ_data_up = requ_data[["Ratings", "Feedback_Code", "Product_Category_Code", "Total_Amount"]]

# Plotting a heatmap to visualize correlation
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")

sns.heatmap(requ_data_up.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title(f'Correlation Between Customer ratings, Feedback, Amt spent and product categories',fontsize=16, fontweight="bold", color="black")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Geographic Analysis:
# How do sales and customer demographics vary by country, state, or city?

# Groping Data based on customer demographics and geogrphy
sales_by_country = df.groupby("Country").agg({"Total_Amount":"sum"})
sales_by_country_cusSeg = df.groupby(["Country","Customer_Segment"]).agg({"Total_Amount":"sum"}).unstack()
sales_by_state_gender = df.groupby(["State","Gender"]).agg({"Total_Amount":"sum"}).sort_values(by="Total_Amount", ascending=False).head(10).unstack()
sales_by_city_income = df.groupby(["City","Income"]).agg({"Total_Amount":"sum"}).sort_values(by="Total_Amount", ascending=False).head(10).unstack()

# Plotting
fig, ((fig1, fig2), (fig3, fig4)) = plt.subplots(2, 2, figsize=(18, 12))

# Pie chart for Total Sales by Country
fig1.pie(sales_by_country["Total_Amount"], labels=sales_by_country.index, autopct='%1.1f%%', startangle=140)
fig1.set_title('Total Sales Percentage by Country', fontsize=16, fontweight='bold')

# Bar chart for Sales by Country and Customer Segments
sales_by_country_cusSeg.plot(kind="bar", ax=fig2, stacked=True, color=["orange", "magenta","grey"], alpha=0.8)
fig2.set_xlabel("Country", fontsize=14, fontweight="bold")
fig2.set_ylabel("Sales (in million)", fontsize=14, fontweight="bold")
fig2.set_title("Sales by Country and Customer Segments", fontsize=16, fontweight="bold")
fig2.legend(["Premium", "Regular"], loc='upper right')

# Bar chart for Sales by State and Gender
sales_by_state_gender.plot(kind='bar', ax=fig3, color=['blue', 'orange'], alpha=0.8)
fig3.set_xlabel('State', fontsize=14, fontweight='bold')
fig3.set_ylabel('Sales (in million)', fontsize=14, fontweight='bold')
fig3.set_title('Top 10 States with Sales by Gender', fontsize=16, fontweight='bold')
fig3.legend(['Female', 'Male'], loc='upper right')
fig3.grid(axis='y', linestyle='--', alpha=0.7)

# Bar chart for Sales by City and Income
sales_by_city_income.plot(kind='bar', ax=fig4, color=['green', 'red', "violet"], alpha=0.8)
fig4.set_xlabel('City', fontsize=14, fontweight='bold')
fig4.set_ylabel('Sales (in million)', fontsize=14, fontweight='bold')
fig4.set_title('Top 10 Cities with Sales by Income', fontsize=16, fontweight='bold')
fig4.legend(['High Income', 'Medium Income', 'Low Income'], loc='upper right')
fig4.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Temporal Analysis:
# How do purchase frequencies vary by month and year?

# Setting up the month in correct order
months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Summarizing Purchase frequencies by month for each year
freq_by_month_2023 = df[df["Year"] == 2023].groupby("Month").size().reindex(months_order, fill_value=0)
freq_by_month_2024 = df[df["Year"] == 2024].groupby("Month").size().reindex(months_order, fill_value=0)

# Plotting lineplot
plt.figure(figsize=(12,8))
sns.lineplot(x= freq_by_month_2023.index, y= freq_by_month_2023.values, ls= "--", linewidth= 2, marker= "o", markersize= 6, label = 2023)
sns.lineplot(x= freq_by_month_2024.index, y= freq_by_month_2024.values, ls= "--", linewidth= 2, marker= "o", markersize= 6, label= 2024)
plt.xlabel("Month", fontsize= 14, fontweight= "bold", color= "black")
plt.ylabel("Purchase Frequency", fontsize= 14, fontweight= "bold", color= "black")
plt.title("Purchase frequency for each month based on year", fontsize= 16, fontweight= "bold", color= "black")
plt.xticks(rotation= 25)
plt.grid(True, ls= "--", alpha= 0.5)
plt.legend(fontsize='large')

plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis:
# Are there any significant correlations between customer demographics (e.g., age, income) and spending behavior?
sig_cor = df[["Age", "Gender", "Income", "Total_Amount"]].copy()

sig_cor["Gender_Code"] = sig_cor["Gender"].astype("category").cat.codes
sig_cor["Income_Code"] = sig_cor["Income"].astype("category").cat.codes
sig_cor_up = sig_cor[["Age", "Gender_Code","Total_Amount", "Income_Code"]]
# Plotting a heatmap to visualize correlation
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")

sns.heatmap(sig_cor_up.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title(f'Correlation Between Age and Total Amount Spent\nCorrelation Coefficient',fontsize=16, fontweight="bold", color="black")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# How do product categories and brands correlate with customer ratings and feedback?
data = df[["Product_Category","Product_Brand","Feedback","Ratings"]].copy()

# Convert categorical variables to numerical using label encoding or one-hot encoding
data["Product_Category_Code"] = data["Product_Category"].astype("category").cat.codes
data["Product_Brand_Code"]    = data["Product_Brand"].astype("category").cat.codes
data["Feedback_Code"]  = data["Feedback"].astype("category").cat.codes

data_up = data[["Product_Category_Code","Product_Brand_Code","Feedback_Code","Ratings"]]

# Plotting a heatmap to visualize correlation
plt.figure(figsize=(10, 8))
sns.set(style="whitegrid")

sns.heatmap(data_up.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},
            xticklabels=["Product Category", "Brand", "Customer Rating", "Feedback"],
            yticklabels=["Product Category", "Brand", "Customer Rating", "Feedback"])

plt.title('Correlation Between Product Categories, Brands, Ratings, and Feedback',
          fontsize=16, fontweight="bold", color="black")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
